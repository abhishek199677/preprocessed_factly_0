```python
import pandas as pd
```

```python
#reading csv and converting it into pandas dataframe

df1 = pd.read_csv(r"C:\Users\putta\Downloads\Table-2.1.csv")
df1.head()
```

```python

 #Specify columns and rows to drop
df1.drop([0,1,2,3,5,36,44,45,46,47,48,49,50,53], inplace = True)

#Drop rows with any NaN values

df1.dropna(axis=0, how='all', inplace=True)
df1.dropna(axis=1, how='all', inplace=True)


df1

```

```python
#to make header row using indexing
df1.columns = df1.iloc[0]
df1.drop([4], inplace = True)
df1

```

```python



# Define a mapping dictionary for column renaming
column_mapping = {
    df1.columns[0]: 'state',
    df1.columns[1]: 'area',
    df1.columns[2]: 'Males',
    df1.columns[3]: 'Females',
    df1.columns[4]: 'Persons',
    df1.columns[5]: 'Rural',
    df1.columns[6]: 'Urban'
}

# Rename columns using the mapping dictionary
df1 = df1.rename(columns=column_mapping)

# Display the updated DataFrame
print(df1)

```

```python


def melt_dataframe(dataframe, id_vars, var_name, value_name):
    melted_df = pd.melt(dataframe, id_vars=id_vars, var_name=var_name, value_name=value_name)
    return melted_df

# Defining id_vars, var_name, and value_name
id_vars = ['state', 'area']
var_name = 'population'
value_name = 'value'

# Calling the function to melt the DataFrame
df1 = melt_dataframe(df1, id_vars, var_name, value_name)

# To Display the melted DataFrame
print(df1)
```

```python


# Insert new columns
df1.insert(4, 'unit', 'Area_in_Square_Kilometers, Population Value in Absolute Number')
df1.insert(5, 'note', '')
df1.insert(0, 'year', 2011)

# Display the DataFrame
print(df1)



```

```python
#adding new data new the required header column using indexing

df1.loc[0, 'note'] = df1.loc[0, 'note'] + "Area figures includes the area under unlawaful occupation of Pakistan and China. The area includes 78,114 sq.km. under illegal occupation of Pakistan, 5,180 sq. km.illegally handed over by Pakistan to China and 37,555 sq.km. under illegal occupation of China."
df1.loc[10, 'note'] = df1.loc[10, 'note'] + 'Area figures includes the area under unlawaful occupation of Pakistan and China. The area includes 78,114 sq.km. under illegal occupation of Pakistan, 5,180 sq. km.illegally handed over by Pakistan to China and 37,555 sq.km. under illegal occupation of China.'
df1
```

```python
#replacing special symbols and characters


def clean_state_names(state):
    state = state.replace('&', 'and')
    state = state.replace('INDIA (1) ++', 'All India')
    state = state.replace('D.and N.Haveli', 'Dadra and Nagar Haveli')
    state = state.replace('A.and N.Islands', 'Andaman and Nicobar Islands')
    state = state.replace('Jammu and Kashmir (1)', 'Jammu and Kashmir')
    state = state.replace('+', '')
    return state

# df1  DataFrame and 'state' column has to be cleaned
df1['state'] = df1['state'].apply(clean_state_names)



#to remove whitespaces
df1[['state', 'unit']] = df1[['state', 'unit']].applymap(lambda x: x.strip() if isinstance(x, str) else x)


display(df1)
```
